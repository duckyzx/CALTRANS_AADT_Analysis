```python
#Importing packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load in the data set
data = pd.read_csv('2020-traffic-volumes.csv')
```

```python
#Visualize the annuual average daily traffic across roads
#Back AADT for first 30 roads in the dataset (DANA POINT, JCT. RTE. to LOS ANGELES, JCT. RTE. 213)
main_df = data.head(30)



back_roads = main_df.DESCRIPTION
back_aadt = main_df.BACK_AADT
plt.bar(back_roads, back_aadt, width = 0.5)
plt.xlabel('Cities')
plt.ylabel('Number of Cars Traveled Anually (BACK AADT)')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Ahead AADT for first 30 roads in the dataset (DANA POINT, JCT. RTE. to LOS ANGELES, JCT. RTE. 213)
ahead_roads = main_df.DESCRIPTION
ahead_aadt = main_df.AHEAD_AADT
plt.bar(ahead_roads, ahead_aadt, width = 0.5)
plt.xlabel('Cities')
plt.ylabel('Number of Cars Traveled Anually (AHEAD AADT)')
plt.xticks(rotation = 90)
plt.show()
```
